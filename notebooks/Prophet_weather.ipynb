```python
a = ["Adilabad","Nizamabad", "Khammam", "Warangal", "Karimnagar"]
for i in a:
  df = pd.read_excel('/content/drive/MyDrive/Climate Change/heatwave/Heat Wave Dataset Cleaned.xlsx',sheet_name=i)
  
  sf = df[['Date','Max Temp (°C)']]
  X = sf[['Date','Max Temp (°C)']]
  y = sf.iloc[:,1]

  train_df = pd.DataFrame()
  train_df['ds'] = pd.to_datetime(X['Date'])
  train_df['y']=y

  model = prophet.Prophet()
  model.fit(train_df)
  future = model.make_future_dataframe(periods=460)
  forecast = model.predict(future)

  forecast = forecast.set_index('ds')

  monthly_means = forecast.groupby(pd.Grouper(freq='M')).max()
  monthly_means.reset_index(inplace=True)
  monthly_means['ds'] = pd.to_datetime(monthly_means['ds'])

  # Extract the year and month values from the date column
  monthly_means['year'] = monthly_means['ds'].dt.year
  monthly_means['month'] = monthly_means['ds'].dt.month
  monthly_means['day'] = 1

  # Create a new datetime object representing the first day of the month
  monthly_means['Date'] = pd.to_datetime(monthly_means[['year', 'month', 'day']])

  # Drop the year and month columns
  monthly_means = monthly_means.drop(['year', 'month'], axis=1)
  monthly_means['Predictions'] = monthly_means['yhat']
  monthly_means = monthly_means[(monthly_means['Date'] > '2022-12-01') & (monthly_means['Date'] < '2024-01-01')]

  ttf = monthly_means[['Date','Predictions']]
  ttf.to_csv('/content/drive/MyDrive/Prophet_weather/weather-{}.csv'.format(i),index=False,header=True)
```
